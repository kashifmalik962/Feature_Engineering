# <center> Data Cleaning </center>

```python
data.head(6)
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(data.isnull())
```

```python
data.isnull().sum()
```

```python
data_isnull_percent = data.isnull().sum()/data.shape[0]*100
```

```python
data_isnull_cols = data_isnull_percent[data_isnull_percent > 20].keys()
```

```python
data_isnull_cols
```

```python
data1 = data.drop(columns=data_isnull_cols, axis=1)
```

```python
print(data.shape)
print(data1.shape)
```

```python
data1.head(6)
```

```python
numerical_cols = data1.select_dtypes(include=['int64', 'float64'])
```

```python
numerical_cols.shape
```

```python
categorical_cols = data1.select_dtypes(include=['object'])
```

```python
categorical_cols.shape
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(numerical_cols.isnull())
```

```python
numerical_missing_val = [var for var in numerical_cols if numerical_cols[var].isnull().sum()>0]
```

```python
plt.figure(figsize=(18,12))
for i, var in enumerate(numerical_missing_val):
  plt.subplot(2,2, i+1)
  sns.distplot(numerical_cols[var], bins=20, kde_kws={'linewidth':'5','color':'red'}, label='Distribution')
  plt.legend()
plt.show()
```

```python
mean_df = numerical_cols.fillna(numerical_cols.mean())
```

```python
median_df = numerical_cols.fillna(numerical_cols.median())
```

```python
plt.figure(figsize=(17,10))
for i,var in enumerate(numerical_missing_val):
  plt.subplot(2, 2, i+1)
  sns.distplot(numerical_cols[var], bins=20, kde_kws={'linewidth':3, 'color' : 'red'}, label='Null_Dist')
  sns.distplot(mean_df[var], bins=20, kde_kws={'linewidth':3, 'color': 'green'}, label='Mean_dist')
  sns.distplot(median_df[var], bins=20, kde_kws={'linewidth':3, 'color':'yellow'}, label='Median_dist')
  plt.legend()
plt.show()
```

```python
for i, var in enumerate(numerical_missing_val):
  plt.figure(figsize=(15, 10))
  plt.subplot(3, 1, 1)
  sns.boxplot(numerical_cols[var])
  plt.subplot(3, 1, 2)
  sns.boxplot(mean_df[var])
  plt.subplot(3, 1, 3)
  sns.boxplot(median_df[var])
```
