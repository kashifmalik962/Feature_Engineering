```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('/content/airlines.csv',usecols=['airline','flight','duration','days_left'])
df.head(4)
```

Feature Splitting

```python
df['flight_type'] = df['flight'].str.split('-', expand=True)[0]
```

```python
df.head(4)
```

```python
ohe = OneHotEncoder(sparse=False, dtype=np.int32)
```

```python
df['flight_type'].value_counts()
```

```python
ohe_transform = ohe.fit_transform(df[['flight_type']])
```

```python
ohe_transform_df = pd.DataFrame(ohe_transform, columns=['6.00E','6E','AI','G8','I5','SG','6.00E'])
```

```python
new_df = pd.concat([df,ohe_transform_df],axis=1)
new_df.head(4)
```

```python
new_df.drop(['airline','flight','flight_type'],axis=1, inplace=True)
```

Feature Construction

```python
def myfunc(num):
  if num <= 1:
    return 0
  if num > 1 and num <= 2:
    return 1
  if num > 2 and num <= 3:
    return 2
  if num > 3 and num <= 5:
    return 3
  else:
    return 4
```

```python
myfunc(8)
```

```python
new_df['duration'] = new_df['duration'].apply(myfunc)
```

```python
new_df.head(4)
```

```python
new_df['duration'] = new_df['duration'] + new_df['days_left'] + 1
```
