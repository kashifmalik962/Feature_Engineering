```python
date = pd.read_csv('/content/orders.csv')
```

Extract Date

```python
date.head(4)
```

```python
date.drop(columns=['product_id','city_id','orders'],inplace=True)
```

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
date.head(4)
```

```python
date['date_year'] = date['date'].dt.year
```

```python
date['date_month'] = date['date'].dt.month_name()
```

```python
date['date_day'] = date['date'].dt.day_name()
```

```python
date['date_week'] = date['date'].dt.week
```

```python
date['date_quater'] = date['date'].dt.quarter
```

```python
date['date_semester'] = np.where(date['date_quater'].isin([1,2]),1,2)
```

```python
date.sample(4)
```

```python
import datetime
```

```python
(((today - date['date']).dt.days)/365)*12
```
